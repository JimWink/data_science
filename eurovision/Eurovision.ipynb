```python
import pandas as pd
import matplotlib.pyplot as plt
```

Data taken from https://data.world/datagraver/eurovision-song-contest-scores

```python
df = pd.read_excel("eurovision_song_contest_1975_2016.xlsx")
df = df[df["(semi-) final"] == "f"]
df = df[["Year", "From country", "To country", "Points"]]
df.head()
```

```python
points = df.groupby(by="To country")["Points"].sum()
points = points.sort_values(ascending=False)
points.head()
```

```python
fig, ax = plt.subplots(figsize=(12, 4))

labels = points.index
values = list(points)
X = range(len(points))
plt.xticks(X, labels, rotation="vertical")
ax.xaxis.set_ticks_position("none")
ax.set_title("Total points (Finals) in Eurovision from 1975 to 2016")
ax.bar(X, values);
```
